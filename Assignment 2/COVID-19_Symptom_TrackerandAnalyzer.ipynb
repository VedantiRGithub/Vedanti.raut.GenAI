```python
import random
class Covid19Model:
    def __init__(self, patient_id, fever, cold, shivering, weight_loss):
        self.patient_id = patient_id
        self.fever = fever
        self.cold = cold
        self.shivering = shivering
        self.weight_loss = weight_loss

    def __repr__(self):
        return (f"Patient ID: {self.patient_id}, "
                f"Fever: {self.fever}°C, Cold: {self.cold}/10, "
                f"Shivering: {self.shivering}/10, Weight Loss: {self.weight_loss}kg")

def generate_covid_data(num_patients=100, seed=42):
    random.seed(seed)
    patients = []
    
    for patient_id in range(1, num_patients + 1):
        fever = round(random.uniform(36.0, 41.0), 1)
        cold = random.randint(0, 10)
        shivering = random.randint(0, 10)
        weight_loss = round(random.uniform(0, 15), 1)

        patient = Covid19Model(patient_id, fever, cold, shivering, weight_loss)
        patients.append(patient)
    
    return patients

def sort_by_symptom(patients, symptom):
    return sorted(patients, key=lambda p: getattr(p, symptom))

patients = generate_covid_data()

sort_parameter = 'fever'

sorted_patients = sort_by_symptom(patients, sort_parameter)
sorted_patients
```
